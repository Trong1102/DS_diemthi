# Đọc từ điểm từ file js

```python
import pandas as pd 
import numpy as np 
```

```python
data_1 = pd.read_json('university_diemchuan.js',lines = True)
```

```python
data_1
```

```python
data_1.university_meta[0]
```

```python
ms =pd.DataFrame({'university_code':[],'university_name':[],'url':[]})
```

```python
for i in range (0,len(data_1['university_meta'])):
  for j in range(0,len(data_1['diemchuan_datas'][i])):
    df_new_row = pd.DataFrame({'university_code':[data_1['university_meta'][i]["university_code"]]
                    ,'university_name':[data_1['university_meta'][i]["university_name"]]
                   ,'url':[data_1['university_meta'][i]["url"]]})
    ms = pd.concat([ms,df_new_row],ignore_index = True)
```

```python
ms
```

```python
new_df = pd.DataFrame({'major_code':[],'major_name':[],'subject_group':[],'point':[],'note':[],'year':[]})
```

```python
for i in range(0,len(data_1['diemchuan_datas'])):
    for j in range(0,len(data_1['diemchuan_datas'][i])):
        temp = pd.DataFrame([data_1['diemchuan_datas'][i][j]])
        new_df = pd.concat([new_df,temp],ignore_index = True)
```

```python
new_df
```

```python
final = pd.concat([new_df,ms],axis = 1,ignore_index = False)
```

```python
final
```
